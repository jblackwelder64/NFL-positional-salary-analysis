```python
#Writing a function to run a t-test on the slope parameter for each of our OLS regressions, with null hypothesis 
#H_0: b_1=0. We do this in order to determine whether a given position group's pay is correlated with team success.
from scipy.stats import t
import math

def t_test(b_1, b_0, X, Y):
    n = len(X)
    S2 = 0
    for i in range(len(X)):
        S2 = S2 + (Y[i]-b_0-b_1*X[i])**2
    
    #S2 at this point has the correct value
    sigprime = math.sqrt(S2/(n-2))
    s_x2 = 0
    for i in range(len(X)):
        s_x2 = s_x2 + (X[i] - np.mean(X))**2
    s_x = math.sqrt(s_x2)
    
    #s_x at this point has the correct value
    u1 = s_x*b_1/sigprime
    p_value = 2*t.cdf(x=-abs(u1), df=(n-2))
    
    return p_value
```

```python
#We will try linear regression on one feature at a time.
import matplotlib.pyplot as plt
for key in df.keys():
    pass
    reg = LinearRegression(fit_intercept=False).fit(np.array(df[key]).reshape(-1, 1), percentages)
    print('R^2:', reg.score(np.array(df[key]).reshape(-1, 1), percentages))
    print('Slope:', reg.coef_[0])
    print('p_value:', t_test(reg.coef_[0], reg.intercept_, np.array(df[key]).reshape(-1, 1), percentages))
    predicted = reg.predict(np.array(df[key]).reshape(-1, 1))
    plt.scatter(np.array(df[key]).reshape(-1, 1), percentages, color="black")
    plt.plot(np.array(df[key]).reshape(-1, 1), predicted, color="blue", linewidth=3)

    plt.title(key)
    plt.xticks(())
    plt.yticks(())

    plt.show()
    
sums_array = []
for i in range(len(df)):
    row_array = np.array(df.iloc[i])
    sums_array.append(sum(row_array))

#normalize the sums_array
sums_array = np.array(sums_array)
mean = sums_array.mean()
std = sums_array.std()
for i in range(len(sums_array)):
    sums_array[i] = (sums_array[i]-mean)/std

reg = LinearRegression(fit_intercept=False).fit(sums_array.reshape(-1, 1), percentages)
print('R^2:', reg.score(sums_array.reshape(-1, 1), percentages))
print('Slope:', reg.coef_[0])
print('p_value:', t_test(reg.coef_[0], reg.intercept_, sums_array.reshape(-1, 1), percentages))
predicted = reg.predict(sums_array.reshape(-1, 1))
plt.scatter(sums_array.reshape(-1, 1), percentages, color="black")
plt.plot(sums_array.reshape(-1, 1), predicted, color="blue", linewidth=3)

plt.title('Total Player Payroll')
plt.xticks(())
plt.yticks(())

plt.show()

# np.array(df[key]).reshape(-1,1)
```

```python
"""
Based on the above regression results, we see that salaries for only three out of the ten position groups appear to be
correlated with team win percentage at a significance level of .05: tight ends (p=.023), edge rushers (p=.001), and 
safeties (p=.007). However, even for these position groups, the R^2 values we obtain from OLS regressions
are all less than .04, suggesting that salary information for position groups alone cannot reliably predict a 
team's success.

This is not to say that we have found no interesting results. For one thing, edge rusher salaries appear to be the
strongest predictor of success among all position groups, even beating out total player payroll in terms of variance
explained and significance level. Several position groups which are typically considered vital, most notably
quarterbacks, wide receivers, and running backs, do not appear to be conclusively correlated with team win percentage.
This leads to the following questions: why does the correlation between player salaries and team win percentage
appear to be so weak? Is talent at some positions not heavily correlated with player ability? Or perhaps some position
groups simply have a smaller impact on the field than others?

We will consider these hypotheses below by identifying intermediate variables meant to quantify on-field performance 
for each position group, and then by considering how these variables relate to the position group's
salary and team success.

"""
```

```python
#Will probably delete this
#Let's try a deep learning model (quick model taken from https://keras.io/api/layers/core_layers/dense/)
#Which we train over the proportional data
from keras.models import Sequential 
from keras.layers import Activation, Dense 
import tensorflow as tf
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

"""
function below taken from https://stackoverflow.com/questions/4601373/better-way-to-shuffle-two-numpy-arrays-in
-unison#:~:text=To%20shuffle%20both%20arrays%20simultaneously,shuffle(c)%20.
"""
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

salary_proportions_cut = df[['TE','EDGE','S']].copy()

# salary_proportions_shuffled, percentages_shuffled = unison_shuffled_copies(salary_proportions_cut.to_numpy(), percentages)
salary_proportions_shuffled, percentages_shuffled = unison_shuffled_copies(df.to_numpy(), percentages)

trainX = np.array(salary_proportions_shuffled)[0:225]
trainY = percentages_shuffled[0:225]

testX = np.array(salary_proportions_shuffled)[225:]
testY = percentages_shuffled[225:]
# print(len(trainX))


model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(10,)))
# model.add(tf.keras.layers.BatchNormalization())
# model.add(tf.keras.layers.Dropout(.2))
model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.BatchNormalization())

# Now the model will take as input arrays of shape (None, 16)
# and output arrays of shape (None, 32).
# Note that after the first layer, you don't need to specify
# the size of the input anymore:
model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(.2))

model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.BatchNormalization())
# model.add(tf.keras.layers.Dropout(.2))

model.add(tf.keras.layers.Dense(1))
# model.summary()

model.compile(optimizer="Adam", loss="mse")
model.fit(trainX, trainY, epochs=6)
predicted = model.predict(testX)

print(mean_squared_error(testY, predicted))
print(r2_score(testY, predicted))


# print(predicted, testY)
```
